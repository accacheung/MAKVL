```python
# Note that there are 2 crime reporting systems: 
# SRS - Summary Reporting System (Prior to 2012)
# NIB - National Incident-Based Reporting (From 2013)
```

```python
# SRS - Summary Reporting System (Prior to 2012)
# "Crimes are recorded in a hierarchical fashion. Only the
# most serious crime is counted whenever multiple offenses are committed in a single incident. 
# Given this hierarchy rule," and the fact that many crimes, especially less serious ones,
# go unreported, the crime index necessarily under-represents the true volume of crimes committed."
```

```python
# NIB - National Incident-Based Reporting (From 2013)
# "While the Summary Reporting System data are recorded in a hierarchical fashion 
# based on eight offense types, NIBRS collects information on twentythree different offense 
# categories made up of 47 offenses and allows all reportable offenses within an incident
# to be reported."
```

```python
# DataFrame for Total Crime and Crime Rate
df_crime_sorted_SRSvsNIB = df_crime_sorted[['Year', 'County','SRS_TOTAL','NIB_TOTAL', 'Total Population']]
```

```python
# Calculate Total Crime (SBS + NIB)
df_crime_sorted_SRSvsNIB['Total Crime'] = df_crime_sorted_SRSvsNIB['SRS_TOTAL'] + df_crime_sorted_SRSvsNIB['NIB_TOTAL']
```

```python
# Calculate Crime Rate (Total Crime/ Total Population)
df_crime_sorted_SRSvsNIB['Crime %'] = ((df_crime_sorted_SRSvsNIB['Total Crime'] / df_crime_sorted_SRSvsNIB['Total Population'])*100).round(2)
```

```python
# DataFrame Preview
df_crime_sorted_SRSvsNIB = df_crime_sorted_SRSvsNIB[['Year', 'County','SRS_TOTAL','NIB_TOTAL', 'Total Crime', 'Total Population', 'Crime %']]
df_crime_sorted_SRSvsNIB.head(14)
```

```python
df_crime_sorted_SRSvsNIB.to_csv('df_crime_sorted_SRSvsNIB.csv', encoding = 'utf-8-sig')
```

```python
# Total Crime per Year
Total_Crime = [df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2010, 'Total Crime'].sum(),
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2011, 'Total Crime'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2012, 'Total Crime'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2013, 'Total Crime'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2014, 'Total Crime'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2015, 'Total Crime'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2016, 'Total Crime'].sum()]

Total_Crime_list = np.array(Total_Crime)
```

```python
# Total Population per Year
Total_Pop = [df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2010, 'Total Population'].sum(),
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2011, 'Total Population'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2012, 'Total Population'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2013, 'Total Population'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2014, 'Total Population'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2015, 'Total Population'].sum(), 
                df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['Year'] == 2016, 'Total Population'].sum()]

Total_Pop_list = np.array(Total_Pop)
```

```python
# Total Crime Rate per Year
# Total_Crime = np.array([Total_Crime], dtype=np.float)
# Total_Pop = np.array([Total_Pop], dtype=np.float)
Crime_from_total = (Total_Crime_list/Total_Pop_list)*100
Crime_from_total.round(2)
```

```python
# Plot of Crime Rate Overtime, from 2010 to 2016
years = df_crime_sorted_SRSvsNIB['Year'].unique()
plt.plot(years, Crime_from_total.round(2), color="red", label="Crime %")
plt.title("Crime Rate Overtime, from 2010 to 2016")
plt.ylabel("Crime Rate %")
plt.xlabel("Years")
plt.ylim(0, 15)

# Saving Plot
plt.savefig("Crime Rate Overtime.png")
```

```python
# Plot of Crime Rate Overtime, Adams County, from 2010 to 2016
y_axis = df_crime_sorted_SRSvsNIB.loc[df_crime_sorted_SRSvsNIB['County'] == 'ADAMS', 'Crime %']

years = df_crime_sorted_SRSvsNIB['Year'].unique()
plt.plot(years, y_axis, color="red", label="Crime %")
plt.title("Crime Rate Overtime, from 2010 to 2016")
plt.ylabel("Crime Rate %")
plt.xlabel("Years")
plt.ylim(0, 15)
```
